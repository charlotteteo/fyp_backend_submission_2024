```python
import xml.etree.ElementTree as ET
from nltk.sentiment import SentimentIntensityAnalyzer
def get_top_company_news(query, num_articles=5):
# Define the URL for the Google News API
    url = f"https://news.google.com/rss/search?q={query}&hl=en-US&gl=US&ceid=US:en"

    # Send an HTTP GET request to the API
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the XML response

        root = ET.fromstring(response.text)
        sid = SentimentIntensityAnalyzer()
        # Initialize a list to store news articles
        articles = []

        # Iterate through the XML to extract article titles and links
        for item in root.findall(".//item")[:num_articles]:
            title = item.find("title").text
            link = item.find("link").text
            pub_date = item.find("pubDate").text  
            # summary = get_article_summary(link)
            sentiment_score = sid.polarity_scores(title)["compound"]
            articles.append({"title": title, "link": link,
                            "sentiment": pub_date})
        

        print(articles)

        return articles, item
    else:
        print(
            f"Failed to retrieve news for {query}. Status code: {response.status_code}")
        return []

articles, item = get_top_company_news('AAPL')
```

```python
#  Calculate daily returns
returns = portfolio_ts.reset_index().set_index('index').pct_change().dropna()

# Calculate maximum drawdown
cumulative_returns = (1 + returns).cumprod()
peak_idx = cumulative_returns.cummax()
drawdown = (cumulative_returns - peak_idx) / peak_idx
drawdown_df = pd.DataFrame(drawdown.min(axis=1))

ts_list = []
for x in drawdown_df.columns:
    data_dict = {'dates':list(drawdown_df[x].index), \
    'values':list(drawdown_df[x].values)}
    ts_list.append(data_dict)

ts_list
```

```python
#  Calculate daily returns
returns = portfolio_ts.reset_index().set_index('index').pct_change().dropna()

# Calculate maximum drawdown
cumulative_returns = (1 + returns).cumprod()
peak_idx = cumulative_returns.cummax()
drawdown = (cumulative_returns - peak_idx) / peak_idx

# Calculate downside deviation
target_return = 0.0
downside_returns = returns.copy()
downside_returns[downside_returns > target_return] = 0
downside_deviation = downside_returns.std()

# Create DataFrame for risk metrics
risk_metrics = pd.DataFrame({'Max Drawdown': drawdown.min(axis=1), 'Downside Deviation': downside_deviation})

# Reset index for plotting
risk_metrics.reset_index(inplace=True)

# Melt DataFrame to long format
risk_metrics_melted = risk_metrics.melt(id_vars=['index'], var_name='Metric', value_name='Value')

# Plot line chart for risk metrics
fig = px.line(risk_metrics_melted, x='index', y='Value', color='Metric', title='Risk Metrics Over Time',
              labels={'Value': 'Value', 'Date': 'Date'}, template='plotly_white')

fig.show()
```
